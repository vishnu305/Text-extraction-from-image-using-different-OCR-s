```python
from paddleocr import PaddleOCR, draw_ocr # main OCR dependencies
from matplotlib import pyplot as plt # plot images
import cv2 #opencv
import os # folder directory navigation
```

# Set up the model

```python
# Setup model
ocr_model = PaddleOCR(lang='en')
```

```python
img_path = os.path.join('.', 'prescription2.png')
result = ocr_model.ocr(img_path)
```

```python
for i in result:
    print(i[1][0],end=" ")
# plt.figure(figsize = (18,12))
plt.imshow(cv2.imread("prescription2.png"))
```

```python
result2 = ocr_model.ocr('prescription3.png')
```

```python
for i in result2:
    print(i[1][0],end=" ")
plt.figure(figsize = (10,10))
plt.imshow(cv2.imread("prescription3.png"))
```

```python
boxes = [res[0] for res in result2] # 
texts = [res[1][0] for res in result2]
scores = [res[1][1] for res in result2]
font_path = os.path.join('PaddleOCR', 'doc', 'fonts', 'latin.ttf')

img = cv2.imread('prescription3.png') 
# reorders the color channels
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 

plt.figure(figsize=(15,15))

# draw annotations on image
annotated = draw_ocr(img, boxes, texts, scores, font_path=font_path) 

# show the image using matplotlib
plt.imshow(annotated) 
```
